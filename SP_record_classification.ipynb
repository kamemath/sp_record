## 予測モデル

```python
base_model = tfk.applications.efficientnet_v2.EfficientNetV2B0(include_top=False, input_shape=(244,244,3))

model1 = tfk.Sequential([
    tfkl.Dense(200, activation='relu'),
    tfkl.Dropout(0.2),
    tfkl.Dense(100, activation='relu'),
    tfkl.Dropout(0.2),
    tfkl.Dense(11, activation='softmax')
])

model = tfk.Sequential([
    base_model,
    model1
])
```

```python
model.compile(
    optimizer=tfk.optimizers.Adam(learning_rate=0.001),
    loss=tfk.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=[
        tfk.metrics.SparseCategoricalAccuracy(),
        tfk.metrics.SparseCategoricalCrossentropy()
    ]
)
```
